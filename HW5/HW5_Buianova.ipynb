```python
import csv


gr1 = ['наша', 'левая', 'правая', 'одна']
gr2 = ['российская', 'советская', 'греческая']
gr3 = ['Петербургская']
gr4 = ['внутренняя', 'лицевая', 'брюшная', 'закатная']
gr5 = ['другая']
gr6 = ['надзирающая', 'российская', 'советская', 'грузинская']


all_gr = [gr1, gr2, gr3, gr4, gr5, gr6]
rows = []

with open('D_SET.csv', 'r', encoding='utf-8') as i:

	for line in i:
		# print(line.split(';'))
		print([one.replace('\n', '') for one in line.split(';')])
		arr = [one.replace('\n', '') for one in line.split(';')]
		_arr = []
		for gr in all_gr:
			for word in arr[:2]:
				if word in gr:
					print('да')
					print(all_gr.index(gr)+1)
					_arr.append('1')
					break
				else:
					_arr.append('0')
					break
		print(arr+_arr)
		print('='*10)

		rows.append(arr+_arr)

with open('storona_keywords.csv', 'w', encoding='utf-8') as w:
	out = csv.writer(w, delimiter=',')
	out.writerows(rows)
```
